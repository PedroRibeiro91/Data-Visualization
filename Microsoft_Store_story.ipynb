```python
# import the necessary libraries
import numpy as np # we will probably need some linear algebra, anyway importing numpy won't hurt 
import pandas as pd  # data visualization tools
import matplotlib.pyplot as plt # graphs
import seaborn as sns # and more graphs
```

```python
# lets read our file
df = pd.read_csv(r'\Users\Pedro\Documents\msft.csv') # read our data file 
```

```python
# A small preview to check the structure of our data

df.head()

# Simple structure 6 columns with names: name, rating, how many people rated, category, data of release and price
# At first sight if we had information about downloads would be nice, it would sure provide better insights
```

```python
# we dont know from the head if we are dealing with non-observed data, but just in case we are
df.dropna()
```

```python
# from this we learned that our data is comprised of 5321 products (after dropping non-observed whether they existed or not) 
# now we are ready to take our second look at the data 
# this time lets see how the categories are distribuited

fig_dims = (10,10)
fig, ax = plt.subplots(figsize = fig_dims)
sns.countplot(y= "Category",data=df, order=df.Category.value_counts().index).set_title('Number of Apps per Category') 
```

```python
# If we were to present this information to a costumer what we would say would be something like:
# On the microsoft store you can find more than 700 music apps, close to 700 book apps and close to 700 business apps
# to name what apps has microsoft invested in the most
```

```python
# Now lets check the number of ratings for each category
# Again would be usefull to have some kind of download per category to at least have a feel for if the ratings are significant
```

```python
fig_dims = (25, 15)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x=df["Category"],y=df["No of people Rated"]).set_title('Number of Ratings per Category') 
```

```python
# this was a surprise
# multimedia and design apps appear as second to last category with the least dedicated apps, on the store, but seems their
# user base are the most vocal in terms of rating, follow by social that appeared at 9th  
# the rest of the categories have about the same number of categories 
```

```python
# what we are going to see next should prove interesting
# lets take a look at free apps vs paid apps
# lets make a change to our data and instead of showing the price in the case of paid apps lets just say they are "Not Free"
df.loc[~df["Price"].isin(['Free']),"Price"] = "Not Free"
```

```python
fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x="Price", data=df).set_title('Free vs Not Free')
```

```python
# as we can see a huge majority of apps in the microsoft store are free to download. And its expected because what trends is 
# an app to have payment options within itself
```

```python
# lets get more detailed and see what information can a free vs not free per category graph bing us
```

```python
fig_dims = (30, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x="Category", hue = "Price", data=df).set_title('Free vs Not Free per Category')
```

```python
# what this graph tells us is from what categories do the small percentage of the paid apps come from
# and those are books, business and development tools
```
